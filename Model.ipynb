# **ANN Model**

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Dense(128, input_shape=(D,), activation='relu'),
  tf.keras.layers.Dense(512,activation = 'relu'),
  tf.keras.layers.Dense(1)
])
```

```python
from keras import backend as K

def root_mean_squared_error(y_true, y_pred):
        return K.sqrt(K.mean(K.square(y_pred - y_true), axis=-1)) 
```

```python
model.compile(
    optimizer = "rmsprop", 
    loss = "mse",
    metrics = ['accuracy']
)
```

```python
df_test_v2 = df_test.iloc[:, 0:6]
```

```python
df_test_v2
```

```python
ytest
```

```python
r = model.fit(
    df_train_v2,ytrain,
    validation_data= (df_test_v2,ytest),
    epochs = 400,
    batch_size = 600

)
```

```python
plt.plot(r.history['loss'], label='loss')
```
